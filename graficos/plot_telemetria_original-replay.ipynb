```python
import matplotlib.pyplot as plt
from matplotlib.ticker import ScalarFormatter

plt.cla() # Clear the current axes
plt.clf() # Clear the current figure

def grafico_telemetria(dt_train, dt_test, dt_train_replay, dt_test_replay, title):
    plt.rcParams["mathtext.fontset"] = 'dejavuserif'
    plt.rcParams["axes.titlesize"] = 16
    plt.rcParams["axes.labelsize"] = 16
    plt.rc('font', size=16)
    plt.rc('figure', facecolor='white')

    # making subplots
    fig, ax = plt.subplots(6,2, figsize=(20,24), linewidth=2, edgecolor="black")
    # set data with subplot and plot
    ax[0,0].plot(dt_train['cpu'], color="tab:blue", label='Original Treino')
    ax[0,0].plot(dt_train_replay['cpu'], color="tab:orange", label='Replay Treino')
    ax[0,0].set_ylabel('% CPU' )
    ax[0,0].set_xlabel('Tempo - Amostras a cada 5s')
    #ax[0,0].set_yscale("symlog")
    #ax[0,0].get_yaxis().set_major_formatter(ScalarFormatter())
    ax[0,0].set_facecolor('ghostwhite')
    ax[0,0].grid()
    ax[0,0].legend()
    ax[0,0].set_title("Uso de CPU - Treino", fontweight='bold')

    # set data with subplot and plot
    ax[0,1].plot(dt_test['cpu'], color="tab:blue", label='Original Teste')
    ax[0,1].plot(dt_test_replay['cpu'], color="tab:orange", label='Replay Teste')
    ax[0,1].set_ylabel('% CPU')
    ax[0,1].set_xlabel('Tempo - Amostras a cada 5s')
    #ax[0,1].set_yscale("symlog")
    #ax[0,1].get_yaxis().set_major_formatter(ScalarFormatter())
    ax[0,1].set_facecolor('ghostwhite')
    ax[0,1].grid()
    ax[0,1].legend()
    ax[0,1].set_title("Uso de CPU - Teste", fontweight='bold')


    ax[1,0].plot(dt_train['memory_usage'], color="tab:blue", label='Original Treino')
    ax[1,0].plot(dt_train_replay['memory_usage'], color="tab:orange", label='Replay Treino')
    ax[1,0].set_ylabel('Megabytes')
    ax[1,0].set_xlabel('Tempo - Amostras a cada 5s')
    #ax[1,1].set_yscale("symlog")
    #ax[1,1].get_yaxis().set_major_formatter(ScalarFormatter())
    ax[1,0].set_facecolor('ghostwhite')
    ax[1,0].grid()
    ax[1,0].legend()
    ax[1,0].set_title("Uso de Memória - Treino", fontweight='bold')
    

    ax[1,1].plot(dt_test['memory_usage'], color="tab:blue", label='Original Teste')
    ax[1,1].plot(dt_test_replay['memory_usage'], color="tab:orange", label='Replay Teste')
    ax[1,1].set_ylabel('Megabytes')
    ax[1,1].set_xlabel('Tempo - Amostras a cada 5s')
    #ax[1,1].set_yscale("symlog")
    #ax[1,1].get_yaxis().set_major_formatter(ScalarFormatter())
    ax[1,1].set_facecolor('ghostwhite')
    ax[1,1].grid()
    ax[1,1].legend()
    ax[1,1].set_title("Uso de Memória - Teste", fontweight='bold')


    ax[2,0].plot(dt_train['network_incoming_packets'], color="tab:blue", label='Original Treino')
    ax[2,0].plot(dt_train_replay['network_incoming_packets'], color="tab:orange", label='Replay Treino')
    ax[2,0].set_ylabel('Pacotes/s')
    ax[2,0].set_xlabel('Tempo - Amostras a cada 5s')
    #ax[2,1].set_yscale("symlog")
    #ax[2,1].get_yaxis().set_major_formatter(ScalarFormatter())
    ax[2,0].set_facecolor('ghostwhite')
    ax[2,0].grid()
    ax[2,0].legend()
    ax[2,0].set_title("Pacotes de entrada (network_incoming_packets) - Treino", fontweight='bold')
    

    ax[2,1].plot(dt_test['network_incoming_packets'], color="tab:blue", label='Original Teste')
    ax[2,1].plot(dt_test_replay['network_incoming_packets'], color="tab:orange", label='Replay Teste')
    ax[2,1].set_ylabel('Pacotes/s')
    ax[2,1].set_xlabel('Tempo - Amostras a cada 5s')
    #ax[2,1].set_yscale("symlog")
    #ax[2,1].get_yaxis().set_major_formatter(ScalarFormatter())
    ax[2,1].set_facecolor('ghostwhite')
    ax[2,1].grid()
    ax[2,1].legend()
    ax[2,1].set_title("Pacotes de entrada (network_incoming_packets) - Teste", fontweight='bold')
    

    ax[3,0].plot(dt_train['network_outgoing_packets'], color="tab:blue", label='Original Treino')
    ax[3,0].plot(dt_train_replay['network_outgoing_packets'], color="tab:orange", label='Replay Treino')
    ax[3,0].set_ylabel('Pacotes/s')
    ax[3,0].set_xlabel('Tempo - Amostras a cada 5s')
    #ax[3,1].set_yscale("symlog")
    #ax[3,1].get_yaxis().set_major_formatter(ScalarFormatter())
    ax[3,0].set_facecolor('ghostwhite')
    ax[3,0].grid()
    ax[3,0].legend()
    ax[3,0].set_title("Pacotes de saída (network_outgoing_packets) - Treino", fontweight='bold')
    

    ax[3,1].plot(dt_test['network_outgoing_packets'], color="tab:blue", label='Original Teste')
    ax[3,1].plot(dt_test_replay['network_outgoing_packets'], color="tab:orange", label='Replay Teste')
    ax[3,1].set_ylabel('Pacotes/s')
    ax[3,1].set_xlabel('Tempo - Amostras a cada 5s')
    #ax[3,1].set_yscale("symlog")
    #ax[3,1].get_yaxis().set_major_formatter(ScalarFormatter())
    ax[3,1].set_facecolor('ghostwhite')
    ax[3,1].grid()
    ax[3,1].legend()
    ax[3,1].set_title("Pacotes de saída (network_outgoing_packets) - Teste", fontweight='bold')


    ax[4,0].plot(dt_train['network_incoming_bytes'], color="tab:blue", label='Original Treino')
    ax[4,0].plot(dt_train_replay['network_incoming_bytes'], color="tab:orange", label='Replay Treino')
    ax[4,0].set_ylabel('Pacotes/s')
    ax[4,0].set_xlabel('Tempo - Amostras a cada 5s')
    #ax[4,1].set_yscale("symlog")
    #ax[4,1].get_yaxis().set_major_formatter(ScalarFormatter())
    ax[4,0].set_facecolor('ghostwhite')
    ax[4,0].grid()
    ax[4,0].legend()
    ax[4,0].set_title("Bytes de entrada (network_incoming_bytes) - Treino", fontweight='bold')
    

    ax[4,1].plot(dt_test['network_incoming_bytes'], color="tab:blue", label='Original Teste')
    ax[4,1].plot(dt_test_replay['network_incoming_bytes'], color="tab:orange", label='Replay Teste')
    ax[4,1].set_ylabel('Pacotes/s')
    ax[4,1].set_xlabel('Tempo - Amostras a cada 5s')
    #ax[4,1].set_yscale("symlog")
    #ax[4,1].get_yaxis().set_major_formatter(ScalarFormatter())
    ax[4,1].set_facecolor('ghostwhite')
    ax[4,1].grid()
    ax[4,1].legend()
    ax[4,1].set_title("Bytes de Entrada (network_incoming_bytes) - Teste", fontweight='bold')


    ax[5,0].plot(dt_train['network_outgoing_bytes'], color="tab:blue", label='Original Treino')
    ax[5,0].plot(dt_train_replay['network_outgoing_bytes'], color="tab:orange", label='Replay Treino')
    ax[5,0].set_ylabel('Pacotes/s')
    ax[5,0].set_xlabel('Tempo - Amostras a cada 5s')
    #ax[5,1].set_yscale("symlog")
    #ax[5,1].get_yaxis().set_major_formatter(ScalarFormatter())
    ax[5,0].set_facecolor('ghostwhite')
    ax[5,0].grid()
    ax[5,0].legend()
    ax[5,0].set_title("Bytes de Saída (network_outgoing_bytes) - Treino", fontweight='bold')
    

    ax[5,1].plot(dt_test['network_outgoing_bytes'], color="tab:blue", label='Original Teste')
    ax[5,1].plot(dt_test_replay['network_outgoing_bytes'], color="tab:orange", label='Replay Teste')
    ax[5,1].set_ylabel('Pacotes/s')
    ax[5,1].set_xlabel('Tempo - Amostras a cada 5s')
    #ax[5,1].set_yscale("symlog")
    #ax[5,1].get_yaxis().set_major_formatter(ScalarFormatter())
    ax[5,1].set_facecolor('ghostwhite')
    ax[5,1].grid()
    ax[5,1].legend()
    ax[5,1].set_title("Bytes de Saída (network_outgoing_bytes) - Teste", fontweight='bold')


    # set the spacing between subplots
    # plt.subplots_adjust(left=0.1,
    #                      bottom=0.1, 
    #                      right=0.9, 
    #                      top=0.8, 
    #                      wspace=0.4, 
    #                      hspace=0.4)
    plt.tight_layout(pad=3.0)
    plt.suptitle(title, fontsize=18, y=1.0)
    plt.show()
```
